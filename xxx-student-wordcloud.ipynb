```python
import jutil
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
def student_cloud(stu_id):
    sql = "select advice1 from appraise_course_record where student_id='%s'"%stu_id
    df = jutil.load_pd_df(sql)
    words = set(df['advice1'])
    txt = str(words)
#     txt = str(words).replace('nbsp','')
    wd = WordCloud(font_path="SimHei.ttf",max_font_size=60,width=600,height=400).generate(txt)

    plt.imshow(wd, interpolation='bilinear',shape=(600,480))
    plt.title(stu_id)
    plt.axis("off")
    plt.show()
```

```python
from io import BytesIO
import base64
```

```python
def student_cloud_imgstr(stu_id):
    sql = "select advice1 from appraise_course_record where student_id='%s'"%stu_id
    df = jutil.load_pd_df(sql)
    words = set(df['advice1'])
    txt = str(words)
#     txt = str(words).replace('nbsp','')
    wd = WordCloud(font_path="SimHei.ttf",max_font_size=60,width=600,height=400).generate(txt)
    b = BytesIO()
    wd.to_image().save(b, "PNG")
    img_str = base64.b64encode(b.getvalue())
    return img_str.decode("ascii")
```
