## 这个要做的事是读取single_links分析下prop拦截值取多少

```python
import json
import os,fnmatch
import matplotlib.pyplot as plt    
```

```python
def get_probs(path):
    probs = []
    num = []
    with open(path, 'r') as f:
        l = f.readlines()
        for x in l:
            d = json.loads(x)
            probs.append(d['prob'])
            num.append(d['total'])
    return probs,num
```

```python
path = "/home/jimo/workspace/Git/great-design/relation-data-2015/"
files = fnmatch.filter(os.listdir(path),'*_single_links.txt')
files
```

```python
nums = []
for file in files:
    probs,num = get_probs(os.path.join(path,file))
    plt.scatter(list(range(len(probs))),probs)
    nums.append(num)
plt.show()
```

```python
for num in nums:
    plt.scatter(list(range(len(num))),num)
    plt.ylim((0,10000))
plt.show()
```

### 从上面可知,有很多都是0.0,所以只要舍去0就行了
