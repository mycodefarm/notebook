```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./mark-data/mark.csv')
data.head()
```

```python
y = []
for i in range(6):
    col = data.iloc[:40,i+1]
    y.append(list(col))
# print(y)
```

```python
x = range(len(y[0]))
for i in range(len(y)):
    plt.plot(x,y[i])
plt.show()
```
