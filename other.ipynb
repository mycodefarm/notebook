```python
import jutil
import os
```

```python
sql = "select * from student_record where family_address like '%合川%' or city  like '%合川%'"
stus = jutil.load_pd_df(sql)
stus.head()
```

```python
stus.to_csv(path_or_buf='hechuan_stu.csv')
```

```python
def save_stus(code,grade,file_name):
    if os.path.exists('stus/'+file_name+'.csv') and os.path.exists('stus/'+file_name+'_photos.csv'):
        return
    sql = "select * from view_student_full_info where speciality_code='%s' and grade='%s'" % (code,grade)
    print(sql)
    stu = jutil.load_pd_df(sql)
    if not stu.empty:
        stu.drop('photo',axis=1).to_csv(path_or_buf='stus/'+file_name+'.csv')
        stu[['student_id','name','photo']].to_csv(path_or_buf='stus/'+file_name+'_photos.csv')
```

```python
save_stus('0402','2015','cs_2015_stus')
```

```python
save_stus('0408','2015','sw_2015_stus')
```

```python
save_stus('0408','2014','sw_2014_stus')
```

```python
# 查出所有专业代码
sql = "select DISTINCT speciality_code,speciality_name from speciality_record"
d_codes = jutil.load_pd_df(sql)
d_codes.head()
```

```python
# 保存下meta data
codes = d_codes['speciality_code'].values
names = d_codes.speciality_name.values
meta = [{'code':codes[i],'name':names[i]} for i in range(len(codes))]
```

```python
jutil.dump_obj(obj=meta,path='stus/meta.txt')
```

```python
for code in codes:
    save_stus(code,'2014',code+'_2014_stus')
    save_stus(code,'2015',code+'_2015_stus')
```
